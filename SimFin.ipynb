```python
df.loc["RUBI"]["simId"]
```

```python
#company overview

companyId = df.loc["RUBI"]["simId"]

base_url = "https://simfin.com/api/v1/companies/id/{0}?api-key={1}".format(companyId, YOUR_API_KEY)
r = requests.get(base_url)
r_json = r.json()
print(r_json)
```

```python
df_overview = pd.DataFrame(r_json, index=[0])
df_overview
```

```python
#list of available company statements

companyId = df.loc["RUBI"]["simId"]

base_url = "https://simfin.com/api/v1/companies/id/{0}/statements/list?api-key={1}".format(companyId, YOUR_API_KEY)
r = requests.get(base_url)
r_json = r.json()
print(r_json)
```

```python
#Return data from statements (not standardized)

#stype: "pl" "bs" "cf"
#ptype: "Q1" "Q2" "Q3" "Q4" "H1" "H2" "9M" "FY" "TTM" "TTM-1" "TTM-2"
#fyear: 2015, 2016, 2017, etc

companyId = df.loc["RUBI"]["simId"]

task = {
    "stype":"pl",
    "ptype":"FY",
    "fyear":"2017"
        }


base_url = "https://simfin.com/api/v1/companies/id/{0}/statements/original?api-key={1}".format(companyId, YOUR_API_KEY)
r = requests.get(base_url, task)
r_json = r.json()
print(r_json)
```

```python
df_pl_nonstd = pd.DataFrame(r_json["values"])
df_pl_nonstd
```

```python
#Return data from statements (standardized)

#stype: "pl" "bs" "cf"
#ptype: "Q1" "Q2" "Q3" "Q4" "H1" "H2" "9M" "FY" "TTM" "TTM-1" "TTM-2"
#fyear: 2015, 2016, 2017, etc

companyId = df.loc["RUBI"]["simId"]

task = {
    "stype":"pl",
    "ptype":"FY",
    "fyear":"2017"
        }

base_url = "https://simfin.com/api/v1/companies/id/{0}/statements/standardised?api-key={1}".format(companyId, YOUR_API_KEY)
r = requests.get(base_url, task)
r_json = r.json()
print(r_json)
```
